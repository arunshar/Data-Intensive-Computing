## Activity 3: K-means clustering with Pew Data

### Install libraries

```r
data <- read.csv("June 10-July 12, 2015 - Gaming, Jobs and Broadband - CSV.csv")
```

### Convert into dataframes

```r
data <- as.data.frame(data)
names(data)
```

### Get certain attributes and remove NA's

```r
data1 <- cbind(data$emplnw, data$eminuse, data$inc, data$smart1, data$act87a)
data2 <- na.omit(data1)
```

### Meaning of these extracted attributes

```r
# EMPLNW

# Are you now employed full-time, part-time, retired, or are you not employed for pay? 

# 1 Employed full-time
# 2 Employed part-time
# 3 Retired
# 4 Not employed for pay
# 5 (VOL.) Have own business/self-employed
# 6 (VOL.) Disabled
# 7 (VOL.) Student
# 8 (VOL.) Other
# 98 (VOL.) Don't know
# 99 (VOL.) Refused


# EMINUSE

# Do you use the internet or email, at least occasionally? {PIAL Teen Trend; PIAL May 2013, Library Typology July-Sept. 2013}

# 1	Yes
# 2	No
# 8	(VOL.) Don't know
# 9	(VOL.) Refused

#Inc

# ASK ALL:
# INC Last year -- that is in 2014 -- what was your total family income from all sources, before taxes? Just stop me when I get to the right category... [READ] {Master INC2}
# 1 Less than $10,000
# 2 10 to under $20,000
# 3 20 to under $30,000
# 4 30 to under $40,000
# 5 40 to under $50,000
# 6 50 to under $75,000
# 7 75 to under $100,000
# 8 100 to under $150,000
# 9 $150,000 or more
# 98 (VOL.) Don't know
# 99 (VOL.) Refused

#SMART1
# Some cell phones are called “smartphones” because of certain features they have. Is your cell phone a smartphone such as an iPhone, Android, Blackberry or Windows phone, or are you not sure? {PIAL Trend}
# 1	Yes, smartphone
# 2	No, not a smartphone
# 8	Not sure/Don't know
# 9	(VOL.) Refused

# ACT87a
# CATEGORIES
# 1 Yes, do this
# 2 No, do not do this
# 8 (VOL.) Don't know
# 9 (VOL.) Refused
```

### Kmeans clustering for 3 clusters

```r
set.seed(20)
k3 <- kmeans(data2, 3, iter.max = 10, nstart = 1, algorithm = c("Hartigan-Wong", "Lloyd", "Forgy", "MacQueen"))
```

```r
k3
```

```r
#x <- as.data.frame(k3$centers)
colnames(k3$centers) <-  c("employment","internet_user","income","smartphone","social_networking sites")
k3$centers
```

### Interpretation : Clustering is based on reponses from the survery with 3 centers :
#### Cluster 1: It seems like people who are employed part-time and uses the internet and have income around 100,000 dollar uses smartphone and are socially active through social media
#### Cluster 2 : People who are retired and internet users having pension around 20,000 do not use smartphone (value is closer than 2) but still they are socially active through social media
#### Cluster 3 : People who are retiered and internet user who refused to give their income also have/don't have smartphone are also socially active through social media

### Analysis: Cluster 3 behaviour is ambiguous to one attributes but we hope that this ambiguity will be resolved by increasing the number of clusters. Within cluster sum of squares by cluster is 97.6 % which should be high because it gives the cohesivity of the cluster. 

### K means clustering for 5 clusters

```r
set.seed(20)
k5 <- kmeans(data2, 5, iter.max = 10, nstart = 1, algorithm = c("Hartigan-Wong", "Lloyd", "Forgy", "MacQueen"))
```

```r
k5
```

```r
colnames(k5$centers) <-  c("employment","internet_user","income","smartphone","social_networking sites")
k5$centers
```
