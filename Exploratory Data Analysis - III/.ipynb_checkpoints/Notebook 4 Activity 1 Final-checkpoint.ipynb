## Activity 1: Linear Model

### Install libraries and real excel sheet

```python
library(xlsx)
library(broom)
data1 <- read.xlsx("NHLTop100.xlsx", sheetIndex = 1)
```

### a. After the initial fitting, evaluate the fit by noting the R2 and p values of the fit.

```python
x <- data1$G
y <- data1$A
lmfit <- lm(y ~ x, data = data1)
plot(x, y, xlab = "GOALS", ylab = "ASSISTS", col = "blue")
intercept <- coef(lmfit)[1]
slope <- coef(lmfit)[2]
abline(intercept,slope)
```

#### Rsquare and p value of original model

```python
cbind(glance(lmfit)[1],glance(lmfit)[5])
```

### b. Change the model so that it line is forced through Wayne Gretzky. Note the R2 and p values of the fit.

```python
wayne <- c(894,1963)
wgoals <- wayne[1]
wassist <- wayne[2]
lmfit1 <- lm(I(y-wassist) ~ I(x-wgoals), data = data1)
intercept1 <- coef(lmfit1)[1]
slope1 <- coef(lmfit1)[2]
intercept2 = -wgoals*slope1 + wassist
plot(x, y, xlab = "GOALS", ylab = "ASSISTS", col = "blue")
abline(intercept,slope)
abline(intercept2,slope1, col = "blue")
```

#### Rsquare and pvalue for Wayne Gretzky data

```python
cbind(glance(lmfit1)[1],glance(lmfit1)[5])
```

### c. Add the data for another player Patrick Kane (he is a current player) and force the line through Kane but not Gretzky. Note R2 and p values of the fit.

```python
patrick <- c(285,465)
pgoals <- patrick[1]
passist <- patrick[2]
lmfit2 <- lm(I(y-passist) ~ I(x-pgoals), data = data1)
intercept3 <- coef(lmfit2)[1]
slope2 <- coef(lmfit2)[2]
intercept4 = -pgoals*slope2 + passist
plot(x, y, xlab = "GOALS", ylab = "ASSISTS", col = "blue")
points(pgoals, passist, col = "red")
abline(intercept,slope)
abline(intercept2,slope1, col = "blue")
abline(intercept4,slope2, col = "red")
```

#### Rsquare and p value for new player Patrick

```python
cbind(glance(lmfit2)[1],glance(lmfit2)[5])
```

### d. Make all the model go through (0,0) and create new models. Note R2 and p values of the fit.

#### Make original model pass through origin

```python
lmfit5 <- lm(y ~ x + 0, data = data1)
slope6 <- coef(lmfit5)[1]
intercept6 = -pgoals*slope6 + passist
plot(x, y, xlab = "GOALS", ylab = "ASSISTS", col = "blue")
abline(intercept6,slope6)
```

#### Rsquare and p value when original model passes through origin

```python
cbind(glance(lmfit5)[1],glance(lmfit5)[5])
```

#### Make Wayne Gretzky pass through origin

```python
lmfit3 <- lm(I(y-wassist) ~ I(x-wgoals) + 0, data = data1)
slope4 <- coef(lmfit3)[1]
intercept4 = -wgoals*slope4 + wassist
plot(x, y, xlab = "GOALS", ylab = "ASSISTS", col = "blue")
abline(intercept4,slope4, col = "blue")
```

#### Rsquare and p value when Wayne data is passed through origin

```python
cbind(glance(lmfit3)[1],glance(lmfit3)[5])
```

#### Make Patrick Kane data pass through origin

```python
lmfit4 <- lm(I(y-passist) ~ I(x-pgoals) + 0, data = data1)
slope5 <- coef(lmfit4)[1]
intercept5 = -pgoals*slope5 + passist
plot(x, y, xlab = "GOALS", ylab = "ASSISTS", col = "blue")
points(pgoals, passist, col = "red")
abline(intercept5,slope5, col = "red")
```

#### Rsqure and p value for forcing Patrick towards origin

```python
cbind(glance(lmfit4)[1],glance(lmfit4)[5])
```
