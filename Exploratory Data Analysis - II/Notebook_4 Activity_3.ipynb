## Loading Libraries

```r
library(ggplot2)
library(dplyr)
```

## Reading CSV and converting into DataFrames

```r
csv1 <- read.csv("June 10-July 12, 2015 - Gaming, Jobs and Broadband - CSV.csv")
dataset <- as.data.frame(csv1)
```

## Question 1

### Check count of visiting dating site on the bases of gender

```r
extract1 <- cbind(dataset[9],dataset[50])
```

```r
male_date <- nrow(extract1 %>% filter(date1a == 1 & sex == 1))
female_date <- nrow(extract1 %>% filter(date1a == 1 & sex == 2))
no_comments <- sum(is.na(extract1$date1a))
H <- c(male_date,female_date,no_comments)
barplot(H, names.arg=c("Male", "Female", "No_response"))
```

### Explaination : Male visit more often than female while many preferred not to answer

## Question 2

### Analayse the people usage of Internet and Cable

```r
extract2 <- cbind(dataset[15],dataset[16],dataset[35])
```

```r
internet_cable <- nrow(extract2 %>% filter(home4nw == 1 & cable1 == 1))
nointernet_cable <- nrow(extract2 %>% filter(home4nw == 2 & cable1 == 1))
nointernet_nocable <- nrow(extract2 %>% filter(home4nw == 2 & cable1 == 2))
H1 <- c(internet_cable,nointernet_cable,nointernet_nocable)
barplot(H1, names.arg=c("Internet & Cable", "Internet & No Cable", "No Internet and No Cable"))
```

### Explaination : From the statistics we can conclude that those who use Internet will most likely use Cable TV.

## Question 3

### Check the count of employement involing labour on the basis of gender

```r
extract3 <- cbind(dataset[9],dataset[70],dataset[71],dataset[72])
```

```r
men_labor <- nrow(extract3 %>% filter(sex == 1 & emptype2 == 1))
women_labor <- nrow(extract3 %>% filter(sex == 2 & emptype2 == 1))
H2 <- c(men_labor,women_labor)
barplot(H2, names.arg=c("Men_Labor", "Women_Labour"))
```

### Explaination : From the plots we analyse that women prefer jobs involving less physical labour than men

## Question 4

### Check the political views of the students

```r
extract4 <- cbind(dataset[70],dataset[131])
```

```r
very_conservative <- nrow(extract4 %>% filter(emplnw == 7 & ideo == 1))
conservative <- nrow(extract4 %>% filter(emplnw == 7 & ideo == 2))
moderate <- nrow(extract4 %>% filter(emplnw == 7 & ideo == 3))
liberal <- nrow(extract4 %>% filter(emplnw == 7 & ideo == 4))
H2 <- c(very_conservative,conservative,moderate,liberal)
barplot(H2, names.arg=c("very_conservative","conservative","moderate","liberal"))
```

### Explaination : Most of the students have liberal view followed by moderate, conservative and very_conservative

## Question 5

### Check the count of social networking sites through Internet and Smartphones

```r
extract5 <- cbind(dataset[12],dataset[24],dataset[41])
```

```r
internet_smartphone <- nrow(extract5 %>% filter(eminuse == 1 & smart1 == 1 & act87a == 1))
internet_nosmartphone <- nrow(extract5 %>% filter(eminuse == 1 & smart1 == 2 & act87a == 1))
neither <- nrow(extract5 %>% filter(eminuse == 2 & smart1 == 2 & act87a == 1))
H2 <- c(internet_smartphone,internet_nosmartphone,neither)
barplot(H2, names.arg=c("internet_smartphone","internet_nosmartphone","neither"))
```

### Explaination : From the plot we analyse that people who are in more social networking activity tend to have both broadband and smartphones than only internet or none of them.
