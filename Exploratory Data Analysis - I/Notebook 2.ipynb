## Extracting user information and saving the location

```python
tweets.df <- twListToDF(tweets) # Convert list into dataframe
names <- tweets.df$screenName # Get screenname column from the tweets.df dataframe
info <- lookupUsers(names) # Get user information from the screename
info.df <- twListToDF(info) # Convert the information in the dataframe
names_loc <- info.df$location # Get the user location column from info.df dataframe
#write.csv(loc, file="loc.csv")
```

## Extracting longitute and latittude from given location

```python
l <- geocode(as.character(names_loc[1:2500])) # Extract the geocode as from all the locations of names_loc
```

## Giving them a table format

```python
comb <- paste(l$lon,l$lat) # Combine two columns in a single column
x1 <- table(comb) # get it in the form of a table and get the freqency of the same location
x2 <- as.data.frame(x1) # convert it to datafrom
x3 <- strsplit(as.character(x2$comb)," ") # split the combined column for longitude and latitude
x4 <- do.call(rbind.data.frame,x3) # Creates a data frame from rows
x4$Freq <- x2$Freq # Add frequency column in the original table
names(x4)[1] <- "lon" # Changing column name for convenience
names(x4)[2] <- "lat" # Changing column name for convenience
write.csv(x4, file = "loc1.csv") # Saving file for future use
```
